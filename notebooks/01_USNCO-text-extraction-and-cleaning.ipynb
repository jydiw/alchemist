```python
# https://stackoverflow.com/questions/33073972/
import os
# os.environ['TIKA_SERVER_JAR'] = 'https://repo1.maven.org/maven2/org/apache/tika/tika-server/1.19/tika-server-1.19.jar'
import tika
from tika import parser

import re
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_colwidth', 0)
pd.set_option('display.max_rows', 150)
```

## USNCO Olympiad Exams

```python
filepaths = []

for file in os.listdir('../data/unsco'):
    if 'local' in file:
        filepaths.append('../data/unsco/'+file)
```

```python
problems = []

for file in filepaths:
    name = file[20:-4]
    raw = parser.from_file(file)
    test = re.split('\\n[0-9]+\.\s', raw['content'])
    for i, problem in enumerate(test):
        problem_dict = {}
        problem_dict['filepath'] = name
        problem_dict['number'] = i
        problem_dict['text'] = problem
        problems.append(problem_dict)
        
df = pd.DataFrame(problems, columns=['filepath', 'number', 'text'])
df = df[df['number'] > 0]
print(df.shape)
df.head()
```

```python
def check(df):
    print(df.shape)
    return df.head()
```

```python
df['text'] = df['text'].str.split('\(A\)')
for i in df.index:
    df.loc[i, 'text'] = df.loc[i, 'text'][0].strip()
check(df)
```

```python
df['chars'] = df['text'].str.len()
check(df)
```

```python
df.sort_values(by='chars')
```

```python
df[df['chars'] < 23].sort_values(by='chars', ascending=False)
```

```python
df = df[df['chars'] > 19]
df['text'] = df['text'].str.replace('\n', '')
check(df)
```

```python
df.reset_index(inplace=True, drop=True)
check(df)
```

```python
df.to_csv('../data/usnco_local.csv', index=False)
```

## textbook problems

```python
file = '../data/textbooks/brown-ccs-3.pdf'
raw = parser.from_file(file)
```

```python
raw['content']
```

```python
problems = []
test = re.split(r'\n\s3\.\d{1,3}\s', raw['content'])
for i, problem in enumerate(test):
    problem_dict = {}
    problem_dict['filepath'] = file[18:-4]
    problem_dict['number'] = i
    problem_dict['text'] = problem
    problems.append(problem_dict)
        
problems_df = pd.DataFrame(problems, columns=['filepath', 'number', 'text'])
problems_df.sort_values(by='number')
```

```python
mask = problems_df['number'] > 0
problems_df = problems_df[mask]
problems_df
```

```python
problems_df.to_csv(f'../data/{file[18:-4]}.csv', index=False)
```
